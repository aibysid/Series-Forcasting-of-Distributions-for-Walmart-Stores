                  ** What is in the Dataset ?**

The dataset consists of three four in the .csv format.

1.sell_prices.csv - contains information about the price of products sold per store and date.

2.calendar.csv - contains the dates on which products are sold in yyyy/dd/mm format.

3.sales_train_validation.csv - Contains the historical daily unit sales data per product and store in the format.

4 .sales_train_evaluation.csv - Dataset to perform evaluation
of the trained model.

                  **What has to be done ?**
The Goal of the project is to forcast the sales for [d_1942 -d_1969] These rows form the evaluation set. The rows [d_1914 - d_1941] form the validation set, and the remaining rows form the training set.
